### Трехмерные точки и линии

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
z1 = np.linspace(0, 15, 1000)
y1 = np.cos(z1)
x1 = np.sin(z1)
ax.plot3D(x1, y1, z1, "green")

z2 = 15 * np.random.random(100)
y2 = np.cos(z2) + 0.1 * np.random.random(100)
x2 = np.sin(z2) + 0.1 * np.random.random(100)
ax.scatter3D(x2, y2, z2, c=z2, cmap="Greens")

plt.show()
```

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))


fig = plt.figure()
ax = plt.axes(projection="3d")
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax.contour3D(X, Y, Z, 40, cmap="binary")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

ax.view_init(60, 45)

plt.show()
```

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))


fig = plt.figure()
ax = plt.axes(projection="3d")
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax.scatter3D(X, Y, Z, c=Z, cmap="Greens")
```

```python
# Каркасный
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))


fig = plt.figure()
ax = plt.axes(projection="3d")
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax.plot_wireframe(X,Y,Z)
```

```python
# Поверхностный
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))


fig = plt.figure()
ax = plt.axes(projection="3d")
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax.plot_surface(X, Y, Z, cmap='viridis',edgecolor='None')
ax.set_title('Example')
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
r = np.linspace(0, 6, 20)
theta = np.linspace(-0.9 * np.pi, 0.8 * np.pi, 40)
R, Theta = np.meshgrid(r, theta)
X = r * np.sin(Theta)
Y = r * np.cos(Theta)
Z = f(X, Y)
ax.plot_surface(X, Y, Z, cmap="viridis", rstride=1, cstride=1)


plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
theta = 2 * np.pi + np.random.random(1000)
r = 6 * np.random.random(1000)
x = r * np.sin(theta)
y = r * np.cos(theta)
z = f(x, y)
ax.scatter3D(x, y, z, c=z, cmap="viridis")
ax.plot_trisurf(x, y, z, cmap="viridis")


plt.show()
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=["x", "y"])
print(data.head)
fog = plt.figure()
plt.hist(data["x"], alpha=0.5)
plt.hist(data["y"], alpha=0.5)

fig = plt.figure()
sns.kdeplot(data=data,shade=True)
```

```python
iris = sns.load_dataset("iris")
print(iris.head())

sns.pairplot(iris, hue="species")

plt.show()
```

```python
tips = sns.load_dataset("tips")
print(tips.head())
grid = sns.FacetGrid(tips, col="sex", row="day",hue='time')
grid.map(plt.hist, "total_bill", bins=np.linspace(0, 40, 15))

plt.show()
```

```python
tips = sns.load_dataset("tips")
print(tips.head())
sns.catplot(data=tips,x='day',y='total_bill',kind='box')


plt.show()
```

```python
tips = sns.load_dataset("tips")
print(tips.head())
sns.jointplot(data=tips, x="tip", y="total_bill", kind="hex")


plt.show()
```
