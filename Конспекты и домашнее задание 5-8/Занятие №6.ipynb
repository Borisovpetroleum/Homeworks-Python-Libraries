```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### Гистограмма

```python
rng=np.random.default_rng(1)
data=rng.normal(size=1000)
plt.hist(data,
         bins=30,
         density=True,
         alpha=0.5,
         histtype='step',
         edgecolor='red')
plt.show()
```

```python
rng = np.random.default_rng(1)
data = rng.normal(size=1000)
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)
args = dict(alpha=0.3,
            bins=40)
plt.hist(x1, **args)
plt.hist(x2, **args)
plt.hist(x3, **args)
plt.show()
```

```python
rng = np.random.default_rng(1)
data = rng.normal(size=1000)
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)
print(np.histogram(x1, bins=1))
print(np.histogram(x2, bins=2))
print(np.histogram(x3, bins=40))
```

## Двумерные гистограммы

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = rng.multivariate_normal(mean, cov, 10000).T
plt.hist2d(x, y, bins=30)
cb = plt.colorbar()
cb.set_label("Point in interval")
plt.show()
print(np.histogram2d(x, y, bins=1))
print(np.histogram2d(x, y, bins=40))
```

```python
plt.hexbin(x, y) # Все тоже самое, только шестиугольнички
```

```python
cities = pd.read_csv("./data/california_cities.csv")
lat, lon, pop, area = (
    cities["latd"],
    cities["longd"],
    cities["population_total"],
    cities["area_total_km2"],
)
plt.scatter(lon, lat, c=np.log10(pop), s=area)
plt.xlabel('Широта')
plt.ylabel('Долгота')
plt.colorbar()
plt.clim(3,7) # Установление пределов в colorbare
plt.scatter([],[],c='r',alpha=0.5,s=100,label='100 $km^2$')
plt.scatter([], [], c="r", alpha=0.5, s=300, label="300 $km^2$")
plt.scatter([], [], c="r", alpha=0.5, s=500, label="500 $km^2$")
plt.legend(labelspacing=3,frameon=False)
plt.show()
```

```python
fig, ax = plt.subplots()
lines = []
styles = ["-", "--", "-.", ":"]
x = np.linspace(0, 10, 1000)
for i in range(4):
    lines += ax.plot(x, np.sin(x - i + np.pi / 2), styles[i])
ax.axis = "equal"
ax.legend(lines[:2], ["line1", "line2"], loc="upper right")
leg = mpl.legend.Legend(ax, lines[1:], ["line2", "line3", "line4"], loc="lower left")
ax.add_artist(leg)
plt.show()
```

## Шкалы

```python
x=np.linspace(0,10,1000)
y=np.sin(x)*np.cos(x[:,np.newaxis])
plt.imshow(y,cmap='Blues')
plt.colorbar()
plt.show()
```

#### Карты цветов
- последовательные(градиент)

- дивергентные(два цвета)

- качественные(смешиваются без четкого порядка)

```python
# последовательные(градиент)
plt.imshow(y,cmap='viridis')
```

```python
# дивергентные(два цвета)
plt.imshow(y,cmap='RdBu')
```

```python
# качественные(смешиваются без четкого порядка)
plt.imshow(y,cmap='jet')
```

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(y, cmap="viridis")
plt.colorbar()

plt.subplot(1, 2, 2)
plt.imshow(y, cmap=plt.cm.get_cmap("viridis",6))
plt.colorbar()
plt.clim(-0.25,0.25)
plt.show()
```
