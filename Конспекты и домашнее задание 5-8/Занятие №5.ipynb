### Два способа вывода графиков

- MATLAB-подобный стиль

- в объектно-ориентиованнном стиле

```python
# MATLAB-подобный стиль
plt.figure()
plt.subplot(2, 1, 1)  # 2 строки, 1 колонка, рисуем первый график из 2
plt.plot(x, np.sin(x))
plt.subplot(2, 1, 2)  # 2 строки, 1 колонка, рисуем второй график из 2
plt.plot(x, np.cos(x))
plt.show()
```

```python
# в объектно-ориентиованнном стиле
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
# fig:plot.Figure - контейнер, содержит объекты(СК, тексты, метки), ax:Sxes - система координат - прямоугольник, деления, метки
plt.show()
```

```python
# Изначально было так
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
plt.show()
```

### Задавать цвета линий color
- 'blue'

- 'rgbcmyk' -> 'rg'

- '0.14' - градация серого от 0 до 1

- RRGGBB - '#FF00EE'

- RGB - (1.0 , 0.2, 0.3)

- HTML - 'salmon'


```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x), color="blue")
ax.plot(x, np.sin(x - 1), color="g")
ax.plot(x, np.sin(x - 2), color="0.75")
ax.plot(x, np.sin(x - 3), color="#FF00EE")
ax.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))
ax.plot(x, np.sin(x - 5), color="salmon")
ax.plot(x, np.sin(x - 6))
```

### Стиль линий linestyle

- сплошная '-', 'solid'

- штриховая '--','dashed'

- штрих-пунктирная '-.','dashdot'

- пунктирная ':', 'dotted'

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x), color="blue")
ax.plot(x, np.sin(x - 1), color="y", linestyle="dashed")
ax.plot(x, np.sin(x - 2), color="0.75", linestyle="dashdot")

ax.plot(x, np.sin(x - 6), "--g")  # Совместное указание цвета и стиля линии
```

```python
fix, ax = plt.subplots(4)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.sin(x))
ax[2].plot(x, np.sin(x))
ax[3].plot(x, np.sin(x))

ax[1].set_xlim(-2, 12)
ax[1].set_ylim(-2, 2)

ax[2].set_xlim(12, -2)
ax[2].set_ylim(2, -2)

ax[3].autoscale(tight=True)
```

```python
plt.subplot(3, 1, 1)
plt.plot(x, np.sin(x),'y')
plt.title('Синус')
plt.xlabel('x')
plt.ylabel('sin(x)')

plt.subplot(3, 1, 2)
plt.plot(x, np.cos(x),':g')
label='cos(x)'
plt.plot(x, np.sin(x), "--k")
label = "sin(x)"
plt.title("Синус и косинус")
plt.xlabel('x')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(x, np.cos(x), ":g")
plt.plot(x, np.sin(x), "--k")

plt.title("Синус и косинус")
plt.xlabel("x")
plt.axis('equal')

plt.legend()

plt.subplots_adjust(hspace=2)
plt.show()
```

```python
x = np.linspace(0, 10, 30)
plt.plot(x, np.sin(x), "s", color="green")
plt.plot(x, np.sin(x)+1, ">", color="green")
plt.plot(x, np.sin(x)+2, "^", color="green")
plt.plot(x, np.sin(x)+3, "o", color="green")
plt.show()
```

```python
x = np.linspace(0, 10, 30)
plt.plot(
    x,
    np.sin(x),
    "--p",
    markersize=15,
    linewidth=4,
    markerfacecolor="y",
    markeredgecolor="r",
    markeredgewidth=2,
)

plt.show()
```

```python
# в scatter можно изменять каждую точку отдельно
# plt.scatter(x, np.sin(x), marker="o")
rng = np.random.default_rng(0)
colors = rng.random(30)
sizes = 100 * rng.random(30)
plt.scatter(x, np.sin(x), marker="o", c=colors, s=sizes)
plt.colorbar()

# Если точек больш 1000, то plot предпочтительнее из-за производительности
```

```python
# Отображение погрешностей

x = np.linspace(0, 10, 50)
dy = 0.4
y = np.sin(x) + dy + np.random.rand(50)
plt.errorbar(x, y, yerr=dy, fmt=".k")
plt.fill_between(x,y-dy,y+dy,color='red',alpha=0.2)
plt.show()
```

```python
def f(x, y):
    return np.sin(x) ** 5 + np.cos(20 + x * y) * np.cos(x)


x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)

# plt.contour(X, Y, Z, cmap='RdGy')
# plt.contourf(X, Y, Z, cmap="RdGy")
c=plt.contour(X, Y, Z, color='red')
plt.clabel(c)
plt.imshow(
    Z,
    extent=[0, 5, 0, 5],
    cmap="RdGy",
    interpolation="gaussian",
    origin="lower",
    aspect="equal",
)
plt.colorbar()
plt.show()
```
